```python
#how many passenger survived 
x=df['Survived'].value_counts() 
surv,Non_surv=x[0],x[1] 
print("Survived People = ",surv,"\n","Non Survived People = ",Non_surv)
```

```python
# visualization The ratio between the number of those who survived and those who did not 
#import matplotlib.pyplot as plt
#plt.pie([surv,Non_surv],labels=["Survived People","Non Survived People"])
import seaborn as sns
sns.countplot(x='Survived', data=df)
```

```python
#the most kind of people survived (male of female)
Male=df.loc[df['Sex']=='male','Survived'].sum()
female=df.loc[df['Sex']=='female','Survived'].sum()
print("number of survived male = ",Male,"\n","number of survived female = ",female)
#visualizing the answer
plt.bar(['survived Male','survived female'],[Male,female])
```
